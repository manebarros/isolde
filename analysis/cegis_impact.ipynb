```python
import pandas as pd
import preprocessing as pre
import domain as dom

df = pd.read_csv('/home/mane/code/minsolde/isolde-experiments/data/80b8403f476d5b.csv')
df = pre.typify(df)
df = pre.trim(df, txn_num=(10,10), implementations = ["all", "no_learning"])
df = df[~df['problem'].apply(lambda p: p.pos[0].name =='CC' and p.neg.name =='PC')]
df = pre.merge_rows(df, check_num_measurements=False)
with pd.option_context('display.max_rows', None):
    display(df)
```

```python
import vldb
df['frameworks'] = df['problem'].apply(lambda p: len(p.frameworks()))
data = vldb.dfToTableData(df, txn_lim=5)
data
```

```python
print(vldb.table3Header)
print(vldb.formatTable3(data))
print(r"\bottomrule")
```
